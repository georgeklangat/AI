# 1. Word cloud with **Square layout**

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text ="From Turing’s vision to today’s AI, machines evolved from simple circuits to minds of metal, learning, seeing, and outsmarting us in life’s complex maze. With endless data and evolving ethics, AI’s path unfolds in algorithms, raising questions of its role in shaping our future unknown."

wordcloud = WordCloud(width=400, height=400, background_color='white').generate(text)

plt.figure(figsize=(5,5))
plt.imshow(wordcloud, interpolation= 'bilinear')

# Turning off the axis lines and lables
plt.axis('off')
plt.show()
```

# 2. **Themed Shape** Word Cloud

```python
from PIL import Image
import numpy as np

heart_mask = np.array(Image.open('/content/drive/MyDrive/heart-37458_1280.png'))

wordcloud1 = WordCloud(mask=heart_mask, background_color='white').generate(text)

plt.figure(figsize=(5,5))
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# 3. **Gradient Word** Cloud

```python
import numpy as np

text = "From Turing’s vision to today’s AI, machines evolved from simple circuits to minds of metal, learning, seeing, and outsmarting us in life’s complex maze. With endless data and evolving ethics, AI’s path unfolds in algorithms, raising questions of its role in shaping our future unknown."
def gradient_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "rgb({}, {}, {})".format(0, min(255, font_size), 0)  # Green gradient by font size

wordcloud3 = WordCloud(color_func=gradient_color_func, background_color='white').generate(text)

plt.figure(figsize=(8, 6))
plt.imshow(wordcloud3, interpolation="bilinear")
plt.axis("off")
plt.show()
```
